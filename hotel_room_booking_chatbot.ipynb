```python
from nltk.chat.util import Chat, reflections
from termcolor import colored

pairs = [    # Define conversation pairs (patterns and responses)
    [
        r'(.*)(room|suit|apartment|studio)',

        ["Okay %2, for how many people ?"]
    ],
    [
        r"(.*)(2|3|4|5|6)",

        ["Perfect! Which month?"]
    ],
    [
        r"(.*)(May|June|July|August|September)",

        ["How long you will stay? Type days or weeks with letters:"]
    ],
    [
        r"(.*)(January|February|March|April|October|November|December)(.*)",

        ["Sorry, we are closed at that month."]
    ],
    [
        r"(.*)(one|two|three|four|five|six|seven|eight|ten|days|day|week|weeks)(.*)",

        ["Lovely! We have a last one left, exactly for you!"]
    ],
    [
        r"(.*)(close|open|work|working)(.*)",

        ["Our hotel works only for the summer season. It opens in May and closing in September."]
    ],
    [
        r"(.*)(price|much|cost)",

        ['One room is 200€ per night.\nThe suit is 300€ per night.\nThe apartment is 300€ per night.\n'
                 'The studio is 100€ per night.']
    ],
    [
        r"(.*)(location|find|place|where)",

        ['We are located at Black sea, Bulgaria']
    ],
    [
        r"quit",

        ["Thank you for booking with me. I wish you the best holiday ever!!!"]
    ],
    [
        r"(.*)(book|confirm|done)",

        ["Your room has been booked! Enjoy your stay at 'Sunrise' hotel! Is there anything else I can help you with before your arrival?"]
    ],
    [
        r"(.*)(amenities|facilities|services|offer)",

        ["We offer a range of amenities, including a swimming pool, spa, fitness center, restaurant, and bar."]
    ],
    [
        r"(.*)(parking|car|vehicle)",

        ["Yes, we have free parking available for our guests."]
    ],
    [
        r"(.*)(beach|sea|ocean|water)",

        ["We're located right on the beautiful Black Sea coast, with direct beach access."]
    ],
    [
        r"(.*)(activities|things to do|explore)",

        ["There are numerous activities available nearby, including water sports, hiking, sightseeing, and exploring the local culture."]
    ],
    [
        r"(.*)(restaurant|food|dining)",

        ["Our on-site restaurant serves delicious cuisine, and we also offer room service."]
    ],
    [
        r"(.*)(pets|dogs|cats|animals)",

        "We are a pet-friendly hotel, so feel free to bring your furry friends along!"
    ],
    [
        r"(.*)(cancellation|policy)",

        ["Our cancellation policy is flexible. Please refer to our website for full details."]
    ],
    [
        r"(.*)(recommend|suggestions|tips)",

        ["I'd be happy to recommend some local attractions and activities based on your interests."]
    ],
    [
        r"(.*)(thank you|thanks)",

        ["You're very welcome! It's my pleasure to assist you."]
    ],
    [
        r"(.*)",

        ['You are welcome to "Sunrise" hotel!']
    ]
]

# default message at the start of chat
print(colored("Please type in English language to start a conversation. "
      "Type quit and leave.\n\nHi, I'm hotel 'Sunrise' reception BOT\nHow can I help you?", "cyan"))
# Create ChatBot instance
chat = Chat(pairs, reflections)

# Start conversation
chat.converse() # Engages in continuous conversation until user types 'quit'
```
